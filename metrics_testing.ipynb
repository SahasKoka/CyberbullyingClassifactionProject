```python
#asdkjhaskljdhaskjdh

from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('few_mistral7B_prompt_final').sheet1

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
#print(rows) # Print the rows to inspect the column names

# Convert to a DataFrame and render.
import pandas as pd
column_names = ['Text', 'Annotation', 'oh_label', 'output']

df = pd.DataFrame.from_records(rows[1:], columns=column_names)
#print(rows)

# Print the column names of the DataFrame to verify
print(df.columns.values)
print(df['oh_label'].unique())
print(df['output'].unique())



from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

# # Load the Excel file
# spreadsheet_key = '1_7VWKN_bpPuhIXlU8tIft5s9UJFcqDGdeGALrAewMGg'  # Replace with your actual key
# sheet_id = worksheet.id
# file_path = f'https://docs.google.com/spreadsheets/d/{spreadsheet_key}/export?format=xlsx&gid={sheet_id}'

# # Extract the true labels and predicted labels
# # Make sure 'oh_label' and 'output' are the correct column names in your DataFrame
y_true = df['oh_label'].astype(int) # Replace 'oh_label' with the actual column name for true labels if needed
y_pred = df['output'].astype(int)   # Replace 'output' with the actual column name for predicted labels if needed

print(y_true.unique())
print(y_pred .unique())
# # Calculate the metrics
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# # Print the metrics
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'Accuracy: {accuracy}')
print(f'F-1 Score: {f1}')
```
